#### En primer lugar, se importa el módulo *os* y de *functools* la función *reduce*, la cual se utilizará más adelante.

```python
import os
from functools import reduce
```

#### Luego, se abren en modo lectura los archivos con los se va a trabajar. Primero se obtiene el directorio actual y luego se completa con la ruta correspondiente a los distintos archivos.

```python
ruta = os.getcwd()
file_nombres_1 = open(os.path.join(ruta,"archivos", "nombres_1.txt"),'r')
file_eval1 = open(os.path.join(ruta,"archivos", "eval1.txt"),'r')
file_eval2 = open(os.path.join(ruta,"archivos", "eval2.txt"),'r')
```

####  Los datos se van leyendo por línea y son convertidos a listas, una por cada archivo. <br> Los datos de los archivos con notas son convertidos a enteros para luego poder sumarlos. <br>Además, se remueven de los string los caracteres " , \n ' ".

```python
notas_1 = [int(x.strip(" ,\n'")) for x in file_eval1.readlines()]
notas_2 = [int(x.strip(" ,\n'")) for x in file_eval2.readlines()]
nombres_1 = [x.strip(" ,\n'") for x in file_nombres_1.readlines()]
```

#### Una vez obtenidas las listas con los datos, se cierran los archivos.

```python
file_nombres_1.close()
file_eval1.close()
file_eval2.close()
```

#### Al haber nombres de estudiantes repetidos, se agrega un número al nombre original para distinguirlos.

```python
nombres = []
for i in range(len(nombres_1)):
    nombres.append(str(i+1) + " " + nombres_1[i])
```

#### Se obtiene una lista con la suma de las notas para cada estudiante.

```python
suma = []
for i in range(len(notas_1)):
    suma.append(notas_1[i]+notas_2[i])
```

#### <div class="alert alert-success">Se crea diccionario donde la clave es el nombre del estudiante y el valor es la suma de sus notas.<div/>

```python
D = dict([(nombres[x],suma[x]) for x in range(len(nombres))])
D
```

#### La función *reduce*  reduce los datos de un iterable a un sólo valor. En este caso sumará las notas de los estudiantes.
#### <div class="alert alert-success">El resultado se divide por la cantidad de notas para obtener el promedio.<div/>

```python
promedio = reduce((lambda x, y: x + y), D.values()) / (2*len(D.values()))
promedio
```

#### <div class="alert alert-success">Finalmente, se informan los estudiantes que obtuvieron menos del promedio.<div/>

```python
for x in D.keys():
    if D[x]/2 < promedio:
        print(x)
```
